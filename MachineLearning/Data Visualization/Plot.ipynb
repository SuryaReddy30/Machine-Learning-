# MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
plt.scatter(x,y,c='r')
plt.show()  # if %matplotlib inline is used then no need of using .show()
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test_Scatterplot.png')
```

```python
# plt plot
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
```

```python
# Creating Subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x=np.arange(1,11)
y=5*x+5
plt.title("demo")
plt.xlabel("X_lable")
plt.ylabel("Y_lable")
plt.plot(x,y)
```

```python
np.pi
```

```python
#Compute the x&y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine Wave Form")
plt.plot(x,y,'g')
```

```python
#Subplot()
#Compute the x&y coordinates for points on sine and cosine curve
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.scatter(x,y_sin)
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
```

```python
# Bar Plot

x=np.arange(0,10,2)
y=np.arange(11,21,2)

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

# Box Plot using matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);     #rectangular box plot
```

```python
data
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False);   
```

# Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.1,0,0.1,0) # explodes 1st slice

#Plot
plt.pie(sizes,explode=explode, labels =labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```

# Distribution plots

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# Correlation With Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

Distplot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#countplot
sns.countplot(y='sex',data=df)
```
